# 1: Business Understanding

Um e-commerce do Reino Unido me contratou para agrupar seus clientes com base nas métricas RFV (discutidas abaixo), para então definir estratégias de fidelização. 

RFM:
- R (Recency): Tempo que o cliente realizou a última compra (em dias)
- F (Frequency): Quantidade de compras realizadas pelo cliente
- V (Value): Valor do **total** gasto pelo cliente


O conjunto de dados fornecido contém transações ocorridas entre dezembro de 2010 e dezembro de 2011.

Seu objetivo é clusterizar os clientes com base em suas métricas de RFV.

```python
df_rfv_clip = df_rfv.apply(lambda x: x.clip(upper=x.quantile(.95)))
df_rfv_clip.describe()
```

# Modeling e Model evaluation
## KMeans

```python
kelbow_visualizer(KMeans(), df_rfv_clip_scaled, k=10, timings=False)
```

```python
cluster_metrics = silhouette_score, davies_bouldin_score, calinski_harabasz_score
cluster_metrics_results = []
X = df_rfv_clip_scaled.copy()

for k in range(2,11):
  model = KMeans(n_clusters=k, random_state=0)
  labels = model.fit_predict(X)
  cluster_results_dict = {'k': k}
  cluster_results_dict['inertia'] = model.inertia_
  for metric in cluster_metrics:
    cluster_results_dict[metric.__name__] = metric(X, labels)
  cluster_metrics_results.append(cluster_results_dict)

pd.DataFrame(cluster_metrics_results).set_index('k').style.background_gradient()
```

```python
kmeans = KMeans(4)
kmeans_labels = kmeans.fit_predict(df_rfv_clip_scaled)
```

```python
px.scatter_3d(df_rfv_clip, x='R', y='F', z='V', color=kmeans_labels.astype(str), template='plotly_dark')
```

# Gaussian Mixture

```python
results = []
k_range = range(1,9)
covariance_types = ['full', 'tied', 'diag', 'spherical']
X = df_rfv_clip_scaled.copy()

for n_components in k_range:
  for covariance_type in covariance_types:
    mclust = GaussianMixture(n_components=n_components, n_init=10, warm_start=True, covariance_type=covariance_type)
    mclust.fit(X)
    results.append({
      'bic': mclust.bic(X),
      'n_components': n_components,
      'covariance_type': covariance_type,
    })

results = pd.DataFrame(results)
results.sort_values('bic').head()
```

```python
sns.lineplot(data=results, x='n_components', y='bic', hue='covariance_type');
```

```python
mclust = GaussianMixture(n_components=4, covariance_type='full', n_init=10)
mclust_labels = mclust.fit_predict(df_rfv_clip_scaled)
```

```python
px.scatter_3d(df_rfv_clip, x='R', y='F', z='V', color=mclust_labels.astype(str), template='plotly_dark')
```

# Interpretação dos clusters

```python
centers = pd.DataFrame(kmeans.cluster_centers_, columns=df_rfv_clip_scaled.columns)
```

```python
fig,axes = plt.subplots(nrows=4, figsize=(14,12), sharex=True)

for i,ax in enumerate(axes):
  center = centers.loc[i,:]
  maxPC = 1.01 * center.abs().max()
  colors = ['green' if l > 0 else 'red' for l in center]
  center.plot.bar(ax=ax, color=colors)
  ax.set_ylabel(f'Cluster {i+1}')
  ax.set_ylim(-maxPC, maxPC)
  ax.axhline(color='gray')
  ax.xaxis.set_ticks_position('none') 

plt.xticks(rotation=60, ha='right')
plt.tight_layout()
plt.show()
```

```python
(
  df_rfv_clip.assign(cluster=kmeans_labels)
  .groupby('cluster')
  .mean()
  .transpose()
  .style.background_gradient(cmap='YlOrRd', axis=1)
)
```

Como objetivo temos que transformar esses clientes em compradores frequentes e fidelizados.

3) Criar conteúdo relevante para este público, como blog posts, tutoriais e vídeos, que demonstrem o valor dos seus produtos e auxiliem na tomada de decisão.

#### Cluster 2: Clientes Frequentes e de Alto Valor

Clientes que compram com frequência e gastam valores altos em cada compra, representando uma parcela significativa da receita do e-commerce.
Temos como objetivo fidelizar ainda mais esses clientes valiosos, aumentando o valor do ticket médio e incentivando compras recorrentes.

##### Ações Recomendadas:
1)  Criar um programa de fidelidade VIP exclusivo para este grupo, oferecendo benefícios diferenciados, como acesso antecipado a lançamentos, pré-vendas exclusivas e convites para eventos VIP.
2)  Enviar ofertas personalizadas com base no histórico de compras e interesses do cliente, utilizando ferramentas de análise de dados e segmentação de clientes.
3)  Oferecer um atendimento ao cliente premium para este grupo, com canais de atendimento exclusivos e profissionais altamente qualificados.

#### Cluster 3: Clientes Frequentes e Baixo Valor

São clientes que realizaram compras recentemente, tem uma boa frequencia, mas ainda não demonstram um padrão de compra consistente.
Como objetivo temos que transformar esses clientes em compradores frequentes e fidelizados.

##### Ações Recomendadas:
1) Enviar e-mails personalizados, apresentando a empresa, ofertas de acordo com os produtos já comprados e sugerir outros com base nas compras anteriores, e oferecendo descontos especiais para a próxima compra.
2) Implementar um programa de fidelidade que recompense os clientes por suas compras, oferecendo pontos, descontos progressivos e benefícios exclusivos.

#### Cluster 4: Churn:
Clientes que já compraram na loja mas, não voltaram.
Objetivo é Reativar esses clientes e fazê-los voltar a comprar.

1) Envio de E-mails personalizados, oferecendo descontos especiais e apresentar novos produtos ou promoções
